```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("output.csv", comment="#", names=["iter",
                                                   "i",
                                                   "x","y",
                                                   "z",
                                                   "virions","chemokine",
                                                   "epicell_status","tcell", "tcell_time"])
```

```python
df.head()
```

```python
def get_iter(df, i):
    iter_frame = df.query("iter == @i")
    epicells = np.zeros((max(df["x"])+1, max(df["y"])+1), dtype=int)
    virions = np.zeros((max(df["x"])+1, max(df["y"])+1))
    chemokine = np.zeros((max(df["x"])+1, max(df["y"])+1))
    tcell_x = []
    tcell_y = []
    for index, row in iter_frame.iterrows():
        x = int(row["x"])
        y = int(row["y"])
        if(int(row["tcell"]) == 1):
            tcell_x.append( int(row["x"]) )
            tcell_y.append( int(row["y"]) )
        epicells[x,y] = int(row["epicell_status"])
        if(float(row["virions"]) > 0.0):
            virions[x,y] = np.log(float(row["virions"]))
        else:
            virions[x,y] = 0.0
        chemokine[x,y] = float(row["chemokine"])
    return {"epicells":epicells, "virions":virions, "chemokine":chemokine,
           "tcell_x":tcell_x, "tcell_y":tcell_y}
```

```python
dt = 1
idx = 0
end = 100
dimx = max(df["x"])
dimy = max(df["y"])
for i in range(0,end,dt):
    it = get_iter(df, i)
    plt.figure()
    plt.imshow(it["virions"])
    plt.savefig("anims/virions/{}.png".format(str(idx).zfill(4)))
    print("anims/virions/{}.png".format(str(idx).zfill(4)))
    plt.close()
    plt.figure()
    plt.imshow(it["chemokine"])
    plt.savefig("anims/chemokine/{}.png".format(str(idx).zfill(4)))
    plt.close()
    print("anims/chemokine/{}.png".format(str(idx).zfill(4)))
    plt.figure()
    plt.scatter(it["tcell_x"], it["tcell_y"])
    plt.xlim([0, dimx])
    plt.ylim([0, dimy])
    plt.savefig("anims/tcell/{}.png".format(str(idx).zfill(4)))
    print("anims/tcell/{}.png".format(str(idx).zfill(4)))
    plt.close()
    idx += 1
```
